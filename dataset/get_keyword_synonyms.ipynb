```python
import requests

kws = requests.get('http://localhost:8000/api/all_labels').json()['keywords']
```

```python
len(kws)
kws_dict = {}
for kw in kws:
    kws_dict[kw['name']] = kw['id']
```

```python
synonymList = []
for line in data.splitlines():
    split = line.split('\t')
    kw = split[0]
    synonyms = [s.strip() for s in split[1].split(',')] if len(split) > 1 else []
    for synonym in synonyms:
        if synonym:
            synonymList.append({'synonym': synonym, 'name': kw, 'db_id': kws_dict[kw]})

synonymList
```

```python
import json
with open('keyword_synonyms.json', 'w') as f:
    json.dump(synonymList, f)
```
